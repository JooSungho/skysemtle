```
from pptx import Presentation
from pptx.util import Inches, Pt
from pptx.enum.text import PP_ALIGN

def create_presentation():
    # 프레젠테이션 객체 생성
    prs = Presentation()

    # 1. 표지 슬라이드
    slide_layout = prs.slide_layouts[0] # Title Slide
    slide = prs.slides.add_slide(slide_layout)
    title = slide.shapes.title
    subtitle = slide.placeholders[1]

    title.text = "뇌성마비복지관 통합돌봄 설계 방안"
    subtitle.text = "다학제적 분석과 지혜로운 통합\n(Multidisciplinary Wisdom & Integrated Care)"

    # 2. 주제의 본질 정의 슬라이드
    slide_layout = prs.slide_layouts[1] # Title and Content
    slide = prs.slides.add_slide(slide_layout)
    title = slide.shapes.title
    content = slide.placeholders[1]

    title.text = "1. 주제의 본질 정의"
    tf = content.text_frame
    tf.text = "시설이라는 '섬'에서 지역사회라는 '숲'으로"

    p = tf.add_paragraph()
    p.text = "핵심 목표:"
    p.level = 1
    p = tf.add_paragraph()
    p.text = "뇌성마비 장애인의 복합적 신체 제약과 의사소통의 어려움 고려"
    p.level = 2
    p = tf.add_paragraph()
    p.text = "지역사회 내에서의 '안전한 자립'과 '주도적 삶'의 공존"
    p.level = 2
    p = tf.add_paragraph()
    p.text = "단순 서비스 확장이 아닌 '생태계 조성'에 초점"
    p.level = 1

    # 3. 다학제적 분석 (1) - 의학/심리
    slide = prs.slides.add_slide(slide_layout)
    title = slide.shapes.title
    content = slide.placeholders[1]

    title.text = "2. 다각도 심층 분석 (Human & Tech)"
    tf = content.text_frame
    tf.text = "🧬 의학 및 재활공학적 관점"

    p = tf.add_paragraph()
    p.text = "생애주기별 재활 유지 관리 (Post-Impairment Syndrome 대비)"
    p.level = 1
    p = tf.add_paragraph()
    p.text = "IoT 기반 스마트홈 & AAC(보완대체 의사소통) 기술 인프라 필수"
    p.level = 1

    p = tf.add_paragraph()
    p.text = "\n🧠 심리 및 사회적 관점"
    p = tf.add_paragraph()
    p.text = "'건강한 의존(Interdependence)' 지향"
    p.level = 1
    p = tf.add_paragraph()
    p.text = "복지관의 역할: 서비스 제공자 → 사회적 관계의 코디네이터"
    p.level = 1
    p = tf.add_paragraph()
    p.text = "수혜자에서 '기여자'로의 정체성 전환 지원"
    p.level = 1

    # 4. 다학제적 분석 (2) - 경영/윤리
    slide = prs.slides.add_slide(slide_layout)
    title = slide.shapes.title
    content = slide.placeholders[1]

    title.text = "2. 다각도 심층 분석 (System & Ethics)"
    tf = content.text_frame
    tf.text = "🧱 경영 및 시스템적 관점"

    p = tf.add_paragraph()
    p.text = "플랫폼(Platform)으로서의 복지관"
    p.level = 1
    p = tf.add_paragraph()
    p.text = "분절된 자원(주거, 의료, 이동)의 패키징(Packaging) 및 중개"
    p.level = 1

    p = tf.add_paragraph()
    p.text = "\n⚖️ 윤리 및 철학적 관점"
    p = tf.add_paragraph()
    p.text = "위험을 감수할 권리 (Dignity of Risk)"
    p.level = 1
    p = tf.add_paragraph()
    p.text = "실패할 기회를 박탈하는 보호는 통제에 불과함"
    p.level = 1
    p = tf.add_paragraph()
    p.text = "스스로 선택하고 결과를 감당하도록 지지하는 윤리적 태도"
    p.level = 1

    # 5. 핵심 쟁점과 딜레마
    slide = prs.slides.add_slide(slide_layout)
    title = slide.shapes.title
    content = slide.placeholders[1]

    title.text = "3. 핵심 쟁점과 딜레마"
    tf = content.text_frame
    tf.text = "🛑 과잉 보호(Safety First) vs 🚀 자기 주도성(Self-Determination)"

    p = tf.add_paragraph()
    p.text = "갈등 지점:"
    p.level = 1
    p = tf.add_paragraph()
    p.text = "복지관/보호자: 의료적 특수성과 사고 우려로 통제된 환경 선호"
    p.level = 2
    p = tf.add_paragraph()
    p.text = "당사자: 지역사회에서의 자유로운 생활과 선택권 갈망"
    p.level = 2
    p = tf.add_paragraph()
    p.text = "해결 과제:"
    p.level = 1
    p = tf.add_paragraph()
    p.text = "이 간극을 메우지 못하면 '지역사회에 위치한 또 다른 시설'로 전락"
    p.level = 2

    # 6. 통합적 해결 방안
    slide = prs.slides.add_slide(slide_layout)
    title = slide.shapes.title
    content = slide.placeholders[1]

    title.text = "4. 통합적 해결 방안 (Methods)"
    tf = content.text_frame
    tf.text = "🏛️ 시스템/정책적 접근: 초연결 케어 매니지먼트"

    p = tf.add_paragraph()
    p.text = "스마트 안심 주거 모델 (AI 스피커, 환경제어장치 활용)"
    p.level = 1
    p = tf.add_paragraph()
    p.text = "지역사회 주치의 제도 및 방문 의료 네트워크 구축"
    p.level = 1
    p = tf.add_paragraph()
    p.text = "개인별 예산제(Personal Budget) 도입 시범 운영"
    p.level = 1

    p = tf.add_paragraph()
    p.text = "\n🤝 개인/의식적 접근: 조력된 의사결정"
    p = tf.add_paragraph()
    p.text = "AAC 마을 만들기 (상점, 관공서 의사소통 판 보급)"
    p.level = 1
    p = tf.add_paragraph()
    p.text = "자립 리허설 (단기 체험홈에서 시행착오 시뮬레이션)"
    p.level = 1

    # 7. 지혜의 한 줄 (Insight)
    slide = prs.slides.add_slide(slide_layout)
    title = slide.shapes.title
    content = slide.placeholders[1]

    title.text = "5. 지혜의 한 줄 (Insight)"
    tf = content.text_frame

    p = tf.add_paragraph()
    p.text = "\"진정한 자립이란 홀로 서는 것이 아니라,\n내가 기대어 설 수 있는 적절한 곳을 스스로 선택하는 것이다.\""
    p.font.bold = True
    p.font.size = Pt(24)
    p.alignment = PP_ALIGN.CENTER

    p = tf.add_paragraph()
    p.text = "\n"

    p = tf.add_paragraph()
    p.text = "제안:"
    p.font.bold = True
    p = tf.add_paragraph()
    p.text = "복지관의 역할은 '지붕(Shelter)'에서 '다리(Bridge)'로 변화해야 합니다."
    p.level = 1
    p = tf.add_paragraph()
    p.text = "성공의 척도: \"얼마나 잘 돌보는가?\" → \"얼마나 잘 연결되어 있는가?\""
    p.level = 1

    # 파일 저장
    prs.save('뇌성마비복지관_통합돌봄_설계.pptx')

if __name__ == "__main__":
    create_presentation()
```
